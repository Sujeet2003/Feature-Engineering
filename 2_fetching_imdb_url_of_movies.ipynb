```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/hindi_movies_id_df')
```

```python
df.head()
```

```python
API_KEY = 'e46c5d37f787329f477b6606a4246fce'
def fetch_imdb_id(movie_id):
  # imdb_id = []
  url = f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}&language=en-US'
  response = requests.get(url).json()
  return response['imdb_id']
```

```python
df['imdb_id'] = df['movies_id'].apply(fetch_imdb_id)
```

```python
df.head()
```

```python
df['imdb_url'] = 'https://www.imdb.com/title/' + df['imdb_id']
```
