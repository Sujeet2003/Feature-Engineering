```python
movie_title = []
release_date = []
duration = []
ratings = []
vote_count = []
poster_path = []
meta_score = []
genres = []
origin_country = []
director = []
cast = []
languages = []
overview = []

# Posteman api: apikey=71a67b9f
# sujeet1772003: apikey=d46dd96
# 1hk21cs161: apikey=32da4b94
# sujeetsm: apikey=a089dc88


def get_movies_details(imdb_id):
  url = f"http://www.omdbapi.com/?i={imdb_id}&apikey=d46dd96"

  response = requests.get(url).json()

  # movie title
  try:
    movie_title.append(response['Title'])
  except:
    movie_title.append(np.nan)


  # release date
  try:
    release_date.append(response['Released'])
  except:
    release_date.append(np.nan)

  # duration
  try:
    duration.append(response['Runtime'])
  except:
    duration.append(np.nan)

  # genres
  try:
    genres.append(response['Genre'])
  except:
    genres.append(np.nan)

  # director
  try:
    director.append(response['Director'])
  except:
    director.append(np.nan)

  # cast
  try:
    cast.append(response['Actors'])
  except:
    cast.append(np.nan)

  # overview
  try:
    overview.append(response['Plot'])
  except:
    overview.append(np.nan)

  # spoken languages
  try:
    languages.append(response['Language'])
  except:
    languages.append(np.nan)

  # country of origin
  try:
    origin_country.append(response['Country'])
  except:
    origin_country.append(np.nan)

  # poster path
  try:
    poster_path.append(response['Poster'])
  except:
    poster_path.append(np.nan)

  # meta score
  try:
    meta_score.append(response['Metascore'])
  except:
    meta_score.append(np.nan)

  # ratings
  try:
    ratings.append(response['imdbRating'])
  except:
    ratings.append(np.nan)

  # vote counts
  try:
    vote_count.append(response['imdbVotes'])
  except:
    vote_count.append(np.nan)
```
